```python
import torch
from torch import nn
```

```python
rnn = nn.LSTM(5, 20, 2)
rnn
```

```python
input = torch.randn(5, 1, 5)

print(input)
```

```python
h0 = torch.randn(2, 1, 20)
c0 = torch.randn(2, 1, 20)
h0
```

```python
output, (hn, cn) = rnn(input, (h0, c0))
```

```python
print(output)
```

```python
print(hn)
```

```python
print(cn)
```
